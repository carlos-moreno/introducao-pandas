```python
import pandas as pd
```

```python
url_file = "https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv"
df = pd.read_csv(url_file)
```

```python
df
```

# Verifique se a base de dados possui dados nulos e, caso tenha, realize o tratamento desses dados nulos da forma que achar mais coerente com a situação.

```python
df = df.fillna(0)
df
```

# Os alunos "Alice" e "Carlos", não fazem mais parte da turma. Sendo assim, remova-os da base de dados.

```python
alunos_a_remover = df.query('Nome == "Alice" | Nome == "Carlos"').index
df.drop(alunos_a_remover, axis=0, inplace=True)
```

# Aplique um filtro que selecione apenas os alunos que foram aprovados.

```python
df = df.query('Aprovado == True')
df
```

# Salve o DataFrame que possui apenas os alunos aprovados em um arquivo csv chamado "alunos_aprovados.csv".

```python
df.to_csv("alunos_aprovados.csv", index=False)
pd.read_csv("alunos_aprovados.csv")
```

# Ao conferir as notas dos alunos aprovados, notamos que algumas notas estavam incorretas. As alunas que tiraram nota 7.0, na verdade, tinham um ponto extra que não foi contabilizado. Sendo assim, substitua as notas 7.0 da base de dados por 8.0.

```python
df.replace(7.0, 8.0, inplace=True)
df.to_csv("alunos_aprovados.csv", index=False)
pd.read_csv("alunos_aprovados.csv")
```
